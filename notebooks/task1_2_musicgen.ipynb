```python
import torch
import torchaudio
import time
from transformers import MusicgenForConditionalGeneration, MusicgenProcessor

device = "cuda" if torch.cuda.is_available() else "cpu"

processor = MusicgenProcessor.from_pretrained("facebook/musicgen-small")
model = MusicgenForConditionalGeneration.from_pretrained(
    "facebook/musicgen-small"
).to(device)

def generate_music(prompt, filename):
    start = time.time()

    inputs = processor(
        text=[prompt],
        padding=True,
        return_tensors="pt"
    ).to(device)

    audio = model.generate(**inputs, max_new_tokens=1024)

    torchaudio.save(
        filename,
        audio[0].cpu(),
        sample_rate=32000
    )

    print(filename, "generated in", round(time.time() - start, 2), "sec")
```

```python
prompts = [
    ("upbeat_happy_pop.wav", "upbeat happy pop music"),
    ("sad_piano.wav", "sad slow piano melody"),
    ("edm_energy.wav", "energetic electronic dance music"),
    ("ambient_calm.wav", "calm peaceful ambient sounds"),
    ("romantic_guitar.wav", "romantic acoustic guitar"),
    ("orchestral_drama.wav", "intense dramatic orchestral"),
    ("funk_groove.wav", "groovy funk bass"),
    ("dark_atmosphere.wav", "mysterious dark atmospheric")
]

for file, prompt in prompts:
    generate_music(prompt, file)
```
